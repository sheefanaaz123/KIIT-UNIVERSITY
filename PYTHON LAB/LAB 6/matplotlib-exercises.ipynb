There's another way to create plots with Matplotlib, it's known as the object-orientated (OO) method. Let's try it.

```python
# Create a plot using plt.subplots()
import numpy as np
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Plot in each subplot
axs[0, 0].plot(x, y, color='b')
axs[0, 0].set_title('Subplot 1')

axs[0, 1].plot(x, -y, color='g')
axs[0, 1].set_title('Subplot 2')

axs[1, 0].plot(x, np.cos(x), color='r')
axs[1, 0].set_title('Subplot 3')

axs[1, 1].plot(x, np.tan(x), color='purple')
axs[1, 1].set_title('Subplot 4')

# Add overall title
fig.suptitle('Multiple Subplots')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a plot using plt.subplots() and then add X & y on the axes

X = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * X)

# Create a figure with a single subplot
fig, ax = plt.subplots()

# Plot the data
ax.plot(X, y, linewidth=2.0, label='y = 4 + 2sin(2x)')
ax.set_title('Plot of y = 4 + 2sin(2x)')
ax.set_xlabel('X values')
ax.set_ylabel('y values')
ax.grid(True)

# Add X and y labels to the axes
ax.annotate('X', xy=(9, 3.5), fontsize=12, color='red')
ax.annotate('y', xy=(0.5, 7), fontsize=12, color='blue')

# Show the legend
ax.legend()

# Show the plot
plt.show()

```

Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers 

car_sales['Price'] = car_sales['Price'].str.replace('[^\d.]', '').str[:-2].astype(float)

print(car_sales)
```

```python
# Add a column called 'Total Sales' which cumulatively adds the 'Price' column
from datetime import datetime, timedelta

car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Add a column called 'Sale Date' which lists a series of successive dates starting from today
today = datetime.today()
date_range = pd.date_range(today, periods=len(car_sales))
car_sales['Sale Date'] = date_range

# View the car_sales DataFrame
print(car_sales)
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column

import matplotlib.pyplot as plt

# Assuming you've already created the car_sales DataFrame with 'Sale Date' and 'Total Sales' columns

# Plot 'Sale Date' versus 'Total Sales'
car_sales.plot(x='Sale Date', y='Total Sales', marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Sale Date')
plt.ylabel('Total Sales')
plt.title('Total Sales Over Time')

# Display the plot
plt.show()

```

```python
# Convert the 'Price' column to the integers


# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function

# Assuming you have the car_sales DataFrame with the 'Price' and 'Odometer (KM)' columns

# Convert 'Price' column to integers
car_sales['Price'] = car_sales['Price'].astype(int)

# Create a scatter plot of 'Odometer (KM)' and 'Price'
car_sales.plot.scatter(x='Odometer (KM)', y='Price')

# Add labels and title
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Scatter Plot of Odometer vs. Price')

# Display the plot
plt.show()

```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

Completing this step correctly should result in a figure which looks like this:
<img src="../images/matplotlib-heart-disease-chol-age-plot-cmap-change.png"/>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already imported the heart_disease DataFrame

# Create DataFrame with patients over 50 years old
patients_over_50 = heart_disease[heart_disease['age'] > 50]

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust figure size if needed

# Plot the data as a scatter plot
ax.scatter(range(len(patients_over_50)), patients_over_50['age'], color='blue', cmap='winter')

# Customize the plot
ax.set_title('Scatter Plot of Age for Patients Over 50 Years Old')
ax.set_xlabel('Index')
ax.set_ylabel('Age')

# Add a mean line
mean_age = patients_over_50['age'].mean()
ax.axhline(mean_age, color='red', linestyle='--', label=f'Mean Age: {mean_age:.2f}')
ax.legend()

# Display the plot
plt.show()
```
